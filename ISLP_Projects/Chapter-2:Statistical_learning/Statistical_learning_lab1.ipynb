```python
'''contour plot(): used to represent 3 dimensional data similar to topographical map'''
''' it takes 3 arguments:
           1. A vector of x values(first dimension)
           2. A vector of y values(second dimension)
           3. A matrix whose elements correspond to z value (thirs dimension) for each pair of (x,y).'''
```

```python
fig,ax=py.subplots(figsize=(8,8))
x=np.linspace(-np.pi,np.pi,50)
y=x
f=np.multiply.outer(np.cos(y),1/(1+x**2))
ax.contour(x,y,f);
```

<h4>Boolean Indexing</h4>

```python
keep_rows=np.zeros(arr.shape[0],bool)
keep_rows
```

```python
keep_rows[[1,2]]=True
keep_rows
```

```python
''' keeps_rows when viewed as integer are the same as the values of np.array([0,1,0,1]).'''
''' np.all(): check whether all entries of an array are true'''
np.all(keep_rows==np.array([0,1,1,0])) 
```

```python
''' np.any(): a similar function is used to check whether any entries of an array are true'''
np.any(keep_rows==np.array([0,1,1,0]))
```

```python
print(arr)
arr[np.array([0,1,1,0])] #it retrieves 0 and 1 rows 
```

```python
''' keep_rows retrieves only second and fourth rows of A- i.e.the rows for which the boolean equals TRUE'''
arr[keep_rows]
```

```python
keep_col=np.zeros(arr.shape[1],bool)
keep_col[[0,2]]=True
idx_bool=np.ix_(keep_rows,keep_col)
print(arr)
arr[idx_bool]
```

```python
keep_col[[0,2,3]]=True
idx_bool=np.ix_(keep_rows,keep_col)
print(arr)
arr[idx_bool]
```

```python
''' we can mix a list with an array of booleans in the arguments to np.ix_'''
idx_mixed=np.ix_([0,3], keep_col)
print(arr)
arr[idx_mixed]
```

<h3>String Formating</h3>

```python
rng=np.random.default_rng(1)
A=rng.standard_normal((127,5))
M=rng.choice([0,np.nan],p=[0.8,0.2],size=A.shape)
A+=M
D=pd.DataFrame(A,columns=['food',
                          'bar',
                          'pickle',
                          'snack',
                          'popcorn'])
D
```

```python
for col in D.columns:
    template='columns "{0}" has {1:2%} missing values'
    print(template.format(col,np.isnan(D[col]).mean()))
```
