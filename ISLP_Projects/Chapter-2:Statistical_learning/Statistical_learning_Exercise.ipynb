This dataset provides information about various colleges, including admission statistics, financial aspects, student demographics, and faculty characteristics. Below is the description of each column:

#### Column Descriptions

1. **Private**: Indicates whether the college is private ("Yes") or public ("No").  
2. **Apps**: The number of applications received by the college.  
3. **Accept**: The number of applications accepted.  
4. **Enroll**: The number of students who actually enrolled.  
5. **Top10perc**: The percentage of students in the top 10% of their high school class.  
6. **Top25perc**: The percentage of students in the top 25% of their high school class.  
7. **F.Undergrad**: The number of full-time undergraduate students.  
8. **P.Undergrad**: The number of part-time undergraduate students.  
9. **Outstate**: The tuition fee for out-of-state students.  
10. **Room.Board**: The cost of room and board (housing and meals).  
11. **Books**: The estimated cost of books.  
12. **Personal**: Estimated personal expenses (e.g., travel, entertainment).  
13. **PhD**: The percentage of faculty members with a PhD.  
14. **Terminal**: The percentage of faculty members with the highest degree in their field (e.g., PhD, MFA).  
15. **S.F.Ratio**: The student-to-faculty ratio.  
16. **perc.alumni**: The percentage of alumni who donate to the college.  
17. **Expend**: The instructional expenditure per student.  
18. **Grad.Rate**: The graduation rate (percentage of students who graduate).
